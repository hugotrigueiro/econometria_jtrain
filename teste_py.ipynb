```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(0, 1, 100)
y = np.random.normal(0.5, 2,100)

```

```python
df = pd.DataFrame({"x":x,
                    "y": y})
```

```python
df.head()
```

```python
df.plot.scatter(x="x",y="y",
title = "teste de correlação",
label ="---")
plt.show()
```
